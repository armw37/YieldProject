```python
import pandas as pd
import flair
import re
```

```python
df=pd.read_csv('10Y_tweets.csv', error_bad_lines=False, sep='\t', lineterminator='\r')
```

```python
display(df)
```

```python
df1 = df[['id','date', 'created_at', 'tweet']]
```

```python
display(df1)
```

```python
df1.to_csv('tweets.csv', index=False)
```

```python
sentiment_model = flair.models.TextClassifier.load('en-sentiment')
```

```python
def clean(tweet):
  whitespace = re.compile(r"\s+")
  web_address = re.compile(r"(?i)http(s):\/\/[a-z0-9.~_\-\/]+")
  user = re.compile(r"(?i)@[a-z0-9_]+")

  # we then use the sub method to replace anything matching
  tweet = whitespace.sub(' ', tweet)
  tweet = web_address.sub('', tweet)
  tweet = user.sub('', tweet)

  tweet = re.sub("\s+", " ", tweet)
  return tweet
```

```python
df1 = df1[0:len(df1)-1]
df1['tweet'] = df1['tweet'].apply(clean)
print(len(df1))
```

```python
df1.to_csv('tweets_processed.csv', index=False)
```

```python
sentiments = []
i = 0
for tweet in df1['tweet']:
  print('Processing', i)
  i += 1
  sentence = flair.data.Sentence(tweet)
  sentiment_model.predict(sentence)
  sentiments.append(sentence.labels[0].value)

df1['sentiment'] = sentiments
```
